```python
import pandas as pd
btc = "../requirements/btc.csv"
# crash = "../requirements/Crash.csv"
df_btc = pd.read_csv(btc)
# df_crash = pd.read_csv(crash)
```

```python
# Check for nulls and datatypes
df_btc.info()
```

**Data Cleaning**

```python
# KNN Imputation
from sklearn.experimental import enable_iterative_imputer
from sklearn.impute import KNNImputer, IterativeImputer
imputer_knn = KNNImputer(n_neighbors=5, weights='uniform')
df_btc_imputed_knn = imputer_knn.fit_transform(df_btc.drop(columns='date'))
df_btc_imputed_knn = pd.DataFrame(df_btc_imputed_knn, columns=df_btc.drop(columns='date').columns)


# Multivariate Imputation by Chained Equations (MICE)
imputer_mice = IterativeImputer
df_btc_imputed_mice = imputer_mice().fit_transform(df_btc.drop(columns='date'))
df_btc_imputed_mice = pd.DataFrame(df_btc_imputed_mice, columns=df_btc.drop(columns='date').columns)
```

```python
df_btc_imputed_knn.head()
```

```python
df_btc_imputed_mice.head()
```

**Data Preprocessing**

```python
from sklearn.preprocessing import MinMaxScaler, StandardScaler, normalize, FunctionTransformer
from sklearn.pipeline import Pipeline
from sklearn.utils import shuffle

# Randomize data
shuffle_knn = shuffle(df_btc_imputed_knn, random_state=42)
shuffle_mice = shuffle(df_btc_imputed_mice, random_state=42)

# Pipeline of preprocessing
pipeline = Pipeline([
    ('minmax', MinMaxScaler()), 
    ('normalize', FunctionTransformer(normalize, validate=False)),
    ('standard', StandardScaler())
    ])

knn_imputed_pipeline = pipeline.fit_transform(shuffle_knn)
mice_imputed_pipeline = pipeline.fit_transform(shuffle_mice)
```

```python
df_btc_imputed_knn_transformed = pd.DataFrame(knn_imputed_pipeline, columns=df_btc.drop(columns=['date']).columns)
df_btc_imputed_mice_transformed = pd.DataFrame(mice_imputed_pipeline, columns=df_btc.drop(columns=['date']).columns)
```

```python
print(df_btc_imputed_knn_transformed.iloc[1000])
```

```python
df_btc_imputed_mice_transformed.head()
```

**Train-test Split**

```python
from sklearn.model_selection import train_test_split

X_knn = df_btc_imputed_knn_transformed.drop(columns=['PriceUSD'])
y_knn = df_btc_imputed_knn_transformed['PriceUSD']
X_mice = df_btc_imputed_mice_transformed.drop(columns=['PriceUSD'])
y_mice = df_btc_imputed_mice_transformed['PriceUSD']

X_knn_train, X_knn_test, y_knn_train, y_knn_test = train_test_split(X_knn, y_knn, test_size=0.2, random_state=42)
X_mice_train, X_mice_test, y_mice_train, y_mice_test = train_test_split(X_mice, y_mice, test_size=0.2, random_state=42)
```

**Linear Regression Model**

```python
from sklearn.linear_model import LinearRegression

lr = LinearRegression()
lr_knn_imputed = lr.fit(X_knn_train, y_knn_train)
print("KNN Imputed:", lr_knn_imputed.score(X_knn_test, y_knn_test))


lr = LinearRegression()
lr_mice_imputed = lr.fit(X_mice_train, y_mice_train)
print("MICE Imputed:", lr_mice_imputed.score(X_mice_test, y_mice_test))
```
